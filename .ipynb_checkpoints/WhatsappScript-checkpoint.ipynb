```python
from selenium import webdriver
from time import sleep
import random
from datetime import datetime

import random
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys

#global variables
userCells = []
userNames = []

#directories
chromeOptionsDir = "user-data-dir=YOUR PROFILE SAVE/LOAD DIR HERE(CAN BE ANYTHING)"
chromeWebDriverDir = 'YOUR CHROME WEBDRIVER EXE LOCATION(DOWNLOAD HERE): https://chromedriver.chromium.org/downloads'
website = "https://web.whatsapp.com"

class WhatsappBot:
    def __init__(self):
        
        # Setting these global for Development purposes
        # if making an application, we will have to rebuild with private access and parameter passing
        global userNames
        
        #this part loads saves cookies in the specified folder 
        chrome_options = Options()
        chrome_options.add_argument(chromeOptionsDir)
        self.driver = webdriver.Chrome(executable_path=chromeWebDriverDir,\
                                       options=chrome_options)
      
        #the site you wanna access, in this case whatsapp web
        self.driver.get(website)
        
        sleep(3)
        
        userNames = self.driver.find_elements_by_xpath('.//span[@class = "_3q9s6"]')
        self.userButtons = self.driver.find_elements_by_class_name('_3m_Xw')
          
        
    def SendMessage(self, message, userList):
        counter = 0
        #loops through available list on the chat area
        for elem in userNames:
            if elem.text in userList:
            
                print("Currently Texting: " , elem.text)
                
                #clicks on user as per given names
                self.userButtons[counter]\
                .click()
                
                sleep(1)
                
                #text box section
                inputElement = self.driver.find_element_by_class_name('p3_M1')
                inputElement.click()    
                
                sleep(1)
                inputElement.send_keys(message)
                
                sleep(1)
                self.driver.find_element_by_class_name('_4sWnG')\
                .click()
                lastTextElement = self.driver.find_elements_by_class_name('_1Gy50')
                for txt in lastTextElement:
                    print(txt.text)
                    
            #the user chat button on the right     
            counter += 1
            
#run the script
sendList = ["THE LIST OF PEOPLE YOU'D LIKE TO SEND A MESSAGE TO, STRING SEPARATED BY COMMAS"]
wBot = WhatsappBot()
wBot.SendMessage('Single Message Test', sendList)
```
